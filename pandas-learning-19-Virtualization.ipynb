# Pandas中的可视化

### 1. 基本绘图
这种在Series和DataFrame结构中的可视化操作只是matplotlib库绘图方法的一层简单封装。  
详情参见[DataFrame.plot()函数官方文档](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.plot.html)

```python
%matplotlib inline

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/1/2000',
    periods=10),columns=list('ABCD'))

df.plot(kind='bar'),'\n'
df.plot.bar(stacked=True),'\n'
df.plot.barh(stacked=True)
```

除了默认线图之外，绘图方法允许其他少量绘图样式。这些方法可以作为plot()的kind关键字参数提供。

### 2. Histograms

```python
df = pd.DataFrame({'a':np.random.randn(1000)+1,'b':np.random.randn(1000),
                  'c':np.random.randn(1000)-1},columns=['a','b','c'])

df.plot.hist(bins=20),'\n'
df.hist(bins=20)
```

### 3. Box Plots

```python
df = pd.DataFrame(np.random.rand(10,5),columns=['A','B','C','D','E'])
df.plot.box()
```

### 4. Area plot

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
df.plot.area()
```

### 5. Scatter Plot

```python
df = pd.DataFrame(np.random.rand(50,4),columns=list('abcd'))
df.plot.scatter(x='a',y='b')
```

### 6. Pie Chart

```python
df = pd.DataFrame(3*np.random.rand(4),index=['a','b','c','d'],columns=['x'])
df.plot.pie(subplots=True)
```
