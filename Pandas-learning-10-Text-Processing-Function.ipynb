# Working with Text Data
这一节主要讨论关于基本的Series/Index字符操作，关于DataFrame的字符函数将在后续章节中介绍。
Pandas提供一系列的字符函数，它们便于对字符串数据进行操作，更重要的是，这些函数会忽略（或排除）掉丢失值(NaN)。

```python
import pandas as pd
import numpy as np

s = pd.Series(['Tom','  William Rick','John','Alber@t',np.nan,'1234','SteveSmith'])
print s,'\n'
print s.str.lower(),'\n'
print s.str.upper(),'\n'
print s.str.len(),'\n'
```

```python
import pandas as pd
import numpy as np

s = pd.Series(['TOM',' William Rick','John','Alber@t'])


# strip()函数帮助在Series/index中每一个字符的两侧跳过空格（包括换行）
print 'After Stripping:'
print s.str.strip()

print 'Split Pattern:'
print s.str.split(' '),'\n'
print s.str.cat(sep='_'),'\n'
# pandas使用get_dummies进行one-hot编码，即离散特征的编码
print s.str.get_dummies(),'\n'
print s.str.contains(' '),'\n'
print 'After replacing @ with $:'
print s.str.replace('@','$'),'\n'
print s.str.repeat(2),'\n'
print s.str.count('m'),'\n'
print s.str.startswith('T'),'\n'
print s.str.endswith('n'),'\n'
print s.str.find('e'),'\n'
print s.str.swapcase(),'\n'
print s.str.islower(),'\n'
print s.str.isupper(),'\n'
print s.str.isnumeric(),'\n'
```
